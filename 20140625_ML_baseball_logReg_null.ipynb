## <p>Analyze the data: select a threshold for the model that defines a strategy for which you would place a wager.  For example, one strategy may be to place wagers on all games where the modeled EV is > T (T is the percent yield of a wager). Calculate the EV as a function of selected T for the null hypothesis.  <p>We also need a way to calculate the variance.  The variance of a binary asset like a wager is nearly useless because it has values of either 0 or something > 1.  To get a useful estimate of the variance of a particular strategy, the wagers have to be binned and the variance of the values of bins can tell how well a strategy pays off.

## We need to step back and look at the true null hypothesis: random selection of games.  Can an algorithm outcomes more accurately than a blind draw? <p>Select games and positions with the same frequency for three cases: random favorites, random underdogs, and random all.  Also compare some basic strategies to exploit potential psychological biases regarding extreme values.     <ul><li>Big underdogs: underdogs are not statistically as unlikely to lose as big moneyline suggest</li>         <li>Big favorites: favorites are more likely win than the moneyline suggests because there is a bias against 'little wins'</li> </ul>

```python
testdf=df.loc[testlist,:]

fraction=0.2
binsize=100;

xx=np.arange(100)

favmeancoll={};favnumcoll={};
undmeancoll={};undnumcoll={};
allmeancoll={};allnumcoll={};

for thres in xx:
    
    favmeancoll[thres]=[]
    favnumcoll[thres]=[]
    count = 0;mc=0.0;nc=0;
    for ii in range(len(testdf)):
        count+=1
        if np.random.rand() < fraction:
            nc+=1
            if (df['AwayValueFinal'][testlist[ii]]<df['HomeValueFinal'][testlist[ii]]):
                mc+=df['AwayPayout'][testlist[ii]]-1
            else:
                mc+=df['HomePayout'][testlist[ii]]-1
        if count == binsize:
            favnumcoll[thres].append(nc+0.01)
            favmeancoll[thres].append(mc)
            mc=0.0;nc=0;count=0;
    
    undmeancoll[thres]=[]
    undnumcoll[thres]=[]
    count = 0;mc=0.0;nc=0;
    for ii in range(len(testdf)):
        count+=1
        if np.random.rand() < fraction:
            nc+=1
            if (df['AwayValueFinal'][testlist[ii]]>df['HomeValueFinal'][testlist[ii]]):
                mc+=df['AwayPayout'][testlist[ii]]-1
            else:
                mc+=df['HomePayout'][testlist[ii]]-1
        if count == binsize:
            undnumcoll[thres].append(nc+0.01)
            undmeancoll[thres].append(mc)
            mc=0.0;nc=0;count=0;
    
    allmeancoll[thres]=[]
    allnumcoll[thres]=[]
    count = 0;mc=0.0;nc=0;
    for ii in range(len(testdf)):
        count+=1
        if np.random.rand() < fraction:
            nc+=1
            if (np.random.rand()>0.5):
                mc+=df['AwayPayout'][testlist[ii]]-1
            else:
                mc+=df['HomePayout'][testlist[ii]]-1
        if count == binsize:
            allnumcoll[thres].append(nc+0.01)
            allmeancoll[thres].append(mc)
            mc=0.0;nc=0;count=0;
  
```

```python
        
F3=plt.figure(figsize=(10,6))
plt.title('Null hypothesis: line-based strategy')
plt.subplot(1,2,1)
for ii in range(len(xx)):
    yii=np.mean(np.divide(allmeancoll[xx[ii]],allnumcoll[xx[ii]]))
    fyii=np.mean(np.divide(favmeancoll[xx[ii]],favnumcoll[xx[ii]]))
    uyii=np.mean(np.divide(undmeancoll[xx[ii]],undnumcoll[xx[ii]]))
    plt.plot(xx[ii],yii,'go')
    plt.plot(xx[ii],fyii,'ro')
    plt.plot(xx[ii],uyii,'bo')
    sdii=np.std(np.divide(allmeancoll[xx[ii]],allnumcoll[xx[ii]]))
    fsdii=np.std(np.divide(favmeancoll[xx[ii]],favnumcoll[xx[ii]]))
    usdii=np.std(np.divide(undmeancoll[xx[ii]],undnumcoll[xx[ii]]))
    plt.errorbar(xx[ii],yii,yerr=sdii, ecolor='g', capthick=2)
    plt.errorbar(xx[ii],fyii,yerr=fsdii, ecolor='r', capthick=2)
    plt.errorbar(xx[ii],uyii,yerr=usdii, ecolor='b', capthick=2)
    #plt.xlim([0.95, 2.05])
    plt.legend(['All','Favorites','Underdogs'],ncol=2,loc=2)
    plt.xlabel('Threshold')
    plt.ylabel('Avg. returns per unit wager')
plt.ylim([-0.4,0.4])
plt.subplot(1,2,2)
alls=[];favs=[];unds=[];
for ii in range(len(xx)):
    alls.append(np.mean(np.divide(allmeancoll[xx[ii]],allnumcoll[xx[ii]])))
    favs.append(np.mean(np.divide(favmeancoll[xx[ii]],favnumcoll[xx[ii]])))
    unds.append(np.mean(np.divide(undmeancoll[xx[ii]],undnumcoll[xx[ii]])))
allmean=np.mean(alls);allsd=np.std(alls);    
favmean=np.mean(favs);favsd=np.std(favs);
undmean=np.mean(unds);undsd=np.std(unds);
plt.bar([0,1,2],[allmean, favmean, undmean],
        color=['g','r','b'],
        yerr=[allsd,favsd,undsd],ecolor='k')
plt.setp(plt.gca(), 'xticks',[0.4,1.4,2.4],
         'xticklabels', ['All','Favorites','Underdogs'])



plt.savefig('naive_baseball_AllFavUnd_returns_test_20140625.png')
```
